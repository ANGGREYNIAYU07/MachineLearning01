1. MEMUAT DATA

```python
import pandas as pd

df = pd.read_csv(path + "Data/day.csv")
df.head()
```

```python
df.describe()
```

2. PRE-PROCESSING DATA DAN FEATURE SELECTION

```python
from sklearn.preprocessing import StandardScaler

# Menentukan Variabel Independen (X) dan Dependen (Y)
X_cols = ['season', 'yr', 'mnth', 'holiday', 'weekday', 'workingday',
          'weathersit', 'temp', 'hum', 'windspeed']
y_col = 'cnt'

X = df[X_cols]
y = df[y_col]

# Mengidentifikasi kolom Kategorikal
categorical_cols = ['season', 'yr', 'mnth', 'holiday', 'weekday', 'workingday', 'weathersit']

# Mengidentifikasi kolom Numerik
numeric_cols = ['temp', 'hum', 'windspeed']

# One-Hot Encoding untuk kolom kategorikal
X_encoded = pd.get_dummies(X, columns=categorical_cols, drop_first=True)

# (Normalisasi) untuk kolom numerik
scaler = StandardScaler()
X_encoded[numeric_cols] = scaler.fit_transform(X_encoded[numeric_cols])

print("\nShape data setelah One-Hot Encoding:", X_encoded.shape)
print("Kolom X final:", X_encoded.columns.tolist())
```

3. PEMBAGIAN DATA

```python
from sklearn.model_selection import train_test_split
# Membagi data menjadi Training Set (80%) dan Testing Set (20%)
X_train, X_test, y_train, y_test = train_test_split(
    X_encoded, y, test_size=0.2, random_state=42
)

print(f"\nTraining set size: {X_train.shape[0]} samples")
print(f"Testing set size: {X_test.shape[0]} samples")
```

4. PELATIHAN MODEL

```python
from sklearn.linear_model import LinearRegression

# Inisialisasi dan latih model Regresi Linear
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
import numpy as np

y_pred = model.predict(X_test)

# Evaluasi
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print("\n--- HASIL EVALUASI MODEL REGRESI LINEAR ---")
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
print(f"R-squared (R2): {r2:.4f}")

# Interpretasi sederhana
print("\nKoefisien Model (Pengaruh Fitur terhadap CNT):")
coefficients = pd.Series(model.coef_, index=X_train.columns)
print(coefficients.sort_values(ascending=False).head(10))
```
